```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
image = cv2.imread("/content/terapia-online-brasil.jpg")
```

```python
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(image_rgb)
plt.axis("off")
plt.show()
```

```python
r_hist = cv2.calcHist([image_rgb], [0], None, [256], [0,256])
g_hist = cv2.calcHist([image_rgb], [1], None, [256], [0,256])
b_hist = cv2.calcHist([image_rgb], [2], None, [256], [0,256])
```

```python
plt.figure(figsize=(10, 6))
plt.plot(r_hist, color='red')
plt.plot(g_hist, color='green')
plt.plot(b_hist, color='blue')
plt.title('Histograma de Cores')
plt.xlabel('Intensidade')
plt.ylabel('Frequência')
plt.legend(['Red', 'Green', 'Blue'])
plt.grid(True)
plt.show()
```

```python
mean_color = np.mean(image_rgb, axis=(0, 1))
std_color = np.std(image_rgb, axis=(0, 1))
max_color = np.amax(image_rgb, axis=(0, 1))
min_color = np.amin(image_rgb, axis=(0, 1))
```

```python
print("Estatísticas de cores na imagem:")
print(f"   Média de cores (R, G, B): {mean_color}")
print(f"   Desvio padrão de cores (R, G, B): {std_color}")
print(f"   Valor máximo de cores (R, G, B): {max_color}")
print(f"   Valor mínimo de cores (R, G, B): {min_color}")
```

```python
new_image = np.ones_like(image_rgb) * mean_color.astype(int)
```

```python
plt.imshow(new_image)
plt.axis("off")
plt.title('Imagem Gerada com Médias de Cores')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
cores_primarias = {
    'vermelho': (1, 0, 0),
    'verde': (0, 1, 0),
    'azul': (0, 0, 1)
}
```

```python
n_tons = 5
```

```python
cores_possiveis = []
```

```python
for nome_cor, cor_rgb in cores_primarias.items():
    for i in range(1, n_tons + 1):
        # Multiplicando a cor primária pela intensidade
        nova_cor = tuple(np.array(cor_rgb) * i)
        cores_possiveis.append(nova_cor)
```

```python
cores_matriz = np.zeros((len(cores_possiveis), 1, 3))
```

```python
for i, cor in enumerate(cores_possiveis):
    cores_matriz[i] = cor
```

```python
plt.figure(figsize=(8, 4))
plt.imshow(cores_matriz, aspect='auto')
plt.axis('off')
plt.title('Cores Possíveis')
plt.show()
```
